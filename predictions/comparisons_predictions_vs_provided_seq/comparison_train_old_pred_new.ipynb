```python
def get_commons(x, y):
    arr = []
    for a in x:
        if a in list(y):
            arr.append(a)
    return arr

def plot_venn_diagram(data1, data2, oname='output.png'):
    plt.figure(figsize=(8,8))
    plt.suptitle(oname)
    common =get_commons(data1, data2)
#     print common
    
    AB = len(common)
    Ab = len(data1) - AB
    aB = len(data2) - AB
    
    print "\tAb:", Ab, ", aB:", aB, ", AB:", AB
    s = (
        Ab,  # Ab
        aB,  # aB
        AB,  # AB
    )

    v = venn2(subsets=s, set_labels=('A', 'B'))

    # Subset labels
    v.get_label_by_id('10').set_text('Ab ='+str(Ab))
    
    v.get_label_by_id('01').set_text('aB ='+str(aB))
   
    if AB != 0:
        v.get_label_by_id('11').set_text('AB ='+str(AB))
        v.get_label_by_id('11').set_y(0.15)

    # Subset colors
    v.get_patch_by_id('10').set_color('c')
    v.get_patch_by_id('01').set_color('#993333')
    if AB != 0:
        v.get_patch_by_id('11').set_color('blue')

    # Subset alphas
    v.get_patch_by_id('10').set_alpha(0.4)
    v.get_patch_by_id('01').set_alpha(1.0)
    if AB != 0:
        v.get_patch_by_id('11').set_alpha(0.7)

    # Border styles
    c = venn2_circles(subsets=s, linestyle='solid')
    c[0].set_ls('dashed')  # Line style
    c[0].set_lw(2.0)       # Line width
    
    for text in v.set_labels:
        text.set_fontsize(16)

#     plt.savefig(oname)
    return common
```

```python
def get_b_not_in_a(data1, data2):
    diff = []
    for d2 in data2:
        if d2 not in data1:
            diff.append(d2)
    return diff


def intersect(data1, data2):
    return list(set(data1) & set(data2))


def get_x_y_axis(data):
    x_ax =[]
    y_ax = []
    annotation = []
    for i in range(0, len(data.keys())):
        k = data.keys()[i]
        x_ax.append(i)
        y_ax.append(len(data[k]))
        annotation.append(k.split("_")[i])
    return [x_ax, y_ax, annotation]


def plot_line_plots(data1, data2, prefix='', ylim1=(0, 1000) , ylim2=(0, 1000)):
    f, axarr = plt.subplots(2, figsize=(10,10))
    x1 = get_x_y_axis(data1)[0]
    y1 = get_x_y_axis(data1)[1]
    l1 = get_x_y_axis(data1)[2]
    
    x2 = get_x_y_axis(data2)[0]
    y2= get_x_y_axis(data2)[1]
    l2 = get_x_y_axis(data1)[2]
    
    axarr[0].scatter(x1, y1)
    axarr[0].plot(y1)
    axarr[0].set_title(prefix+' Union')
    axarr[0].set_ylim(ylim1)
    
    axarr[1].scatter(x2, y2)
    axarr[1].plot(y2)
    axarr[1].set_title(prefix+' B_not_A')
    axarr[1].set_ylim(ylim2)
    get_x_y_axis
    
    for a in range(0 , len(x1)):
        axarr[0].annotate(l1[a], xy = (x1[a], y1[a]))
    
    for a in range(0 , len(x2)):
        axarr[1].annotate(l2[a], xy = (x2[a], y2[a]))
        
#     f.subplots_adjust(hspace=0.3)
    
    
def get_unions_bNa(data, iterable=[]):
    d_u = OrderedDict()
    d_i = OrderedDict()
    d_b_not_in_a=OrderedDict()
    
    key = iterable[0]
    for i in range(0, len(iterable)):
        key_o = key
        key = "%s_%s" % (key, iterable[i])
        if i==0:
            d_u[key] = list(set(data[iterable[i]]).union(data[iterable[i]]))
            d_i[key] = intersect(data[iterable[i]], data[iterable[i]])
            d_b_not_in_a[key] = get_b_not_in_a(d_i[key], data[iterable[i]])
        
        else: 
            d_u[key] = list(set(d_u[key_o]).union(data[iterable[i]]))
            d_i[key] = intersect(d_u[key_o], data[iterable[i]])
            d_b_not_in_a[key] = get_b_not_in_a(d_i[key], data[iterable[i]])
        
    return [d_u, d_i, d_b_not_in_a]

    
def analyze_overlap(data, iterable=[], top=0, bottom=0):
    dict_seq_top = OrderedDict()
    dict_seq_bottom = OrderedDict()
    
    for itr in iterable:
        for fn in data:
            if '_'+str(itr)+'_LSTM' in fn:
                df = pd.read_csv(fn, sep='\t', usecols=[0])
                df.columns = ['seq']
                dict_seq_top[itr] = df['seq'][0:top]
                dict_seq_bottom[itr] = df['seq'][-1*(bottom):]
    
    return [get_unions_bNa(dict_seq_top, iterable), get_unions_bNa(dict_seq_bottom, iterable)]
```

```python
def calculate_mean_and_stdev(data, dataset='', type = ''):
    if dataset == 'top':
        ds = 0
    if dataset == 'bottom':
        ds = 1    
    
    if type == 'union':
        tp = 0
    if type == 'B_not_A':
        tp = 2
    
#     print type, tp, dataset, ds
    top_union_average = []
    y_ax = []
    for itr in data:
        temp=[]
        for i in range(0, len(data[itr][ds][tp].keys())):
            k = data[itr][ds][tp].keys()[i]
            temp.append(len(data[itr][ds][tp][k]))
        y_ax.append(temp)
#     print np.mean(y_ax, axis=0)
    return [np.mean(y_ax, axis=0), np.std(y_ax, axis=0)]


def plot_line_plots_mean_stdev(data1, data2, prefix='', ylim1=(0, 1000) , ylim2=(0, 1000)):
    f, axarr = plt.subplots(2, figsize=(6,8))
    mean1 = data1[0]
    error1 = data1[1]
    mean2 = data2[0]
    error2 = data2[1]
    
    x1 = np.arange(0, len(mean1))
    y1 = mean1
    
    x2 = np.arange(0, len(mean2))
    y2= mean2
    
    axarr[0].errorbar(x1, y1, error1, marker='.', linewidth=1, ecolor='red')
#     axarr[0].plot(y1)
    axarr[0].set_title(prefix+' Union')
    axarr[0].set_ylim(ylim1)
    axarr[0].set_xlim((1, 24))
    
    axarr[1].errorbar(x2, y2, error2, marker='.', linewidth=1, ecolor='red')
#     axarr[1].plot(y2)
    axarr[1].set_title(prefix+' B_not_A')
    axarr[1].set_ylim(ylim2)
    axarr[1].set_xlim((1, 24))
    f.subplots_adjust(hspace=0.3)
    

```

```python
no_of_iterations=24
reshuffle = 100
iterable = np.arange(no_of_iterations)
random_list = []

for i in range(reshuffle):
    rand = np.random.permutation(iterable)
    random_list.append(rand)
    print "iteration_%s"%i, rand
    
    
dict_overlap_data_top_bottom = OrderedDict()

# pos_neg_dataset_size = [5, 10]

pos_neg_dataset_size = [100, 250, 500, 1000, 5000, 10000]

for size in pos_neg_dataset_size:
    dict_overlap_data_top_bottom[size] = OrderedDict()
    for i in range(len(workdir_path)):
        wdir = workdir_path[i]
        dict_overlap_data_top_bottom[size][wdir] = OrderedDict()

        prefix = os.path.split(wdir)[1]
        arr_txt_files = glob.glob(os.path.join(wdir, 'all_predict*.txt'))

        for rand in random_list:
            output = analyze_overlap(arr_txt_files, iterable=rand , top=size, bottom=size)
            dict_overlap_data_top_bottom[size][wdir]['_'.join([str(x) for x in rand])] = output
```

```python
def mean_squared_error(y, pred):
    # mean squared error calculation
    return np.mean((y - pred) ** 2)
```
